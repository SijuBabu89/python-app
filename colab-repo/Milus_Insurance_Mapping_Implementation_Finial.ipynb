```python
ins_id = FieldSchema(name = 'id', dtype=DataType.INT64,is_primary =True, max_length=50)
ecw_insurance_name = FieldSchema(name = 'ecw_insurance_name', dtype=DataType.VARCHAR, max_length=200)
eligibility_insurance_attribute_vector = FieldSchema(name = 'eligibility_insurance_attribute_vector', dtype=DataType.FLOAT_VECTOR, dim=384)
```

```python
ins_mapping_schema = CollectionSchema(fields = [ins_id, ecw_insurance_name, eligibility_insurance_attribute_vector], description='Insurance Mapping Collection')
```

```python
collection = Collection(name='ECWEligibilityInsurances', schema=ins_mapping_schema, using = 'default')
```

```python
utility.list_collections()
```

```python
import pandas as pd
df = pd.read_csv("ECWEligibilityMapping - Data.csv")
insurance_dict = {}
ids = []
i=1
mapping_fields = []
ecw_insurance = []
insurance_vector = []

for row, column in df.iterrows():
  mapping_fields.append(str(column['Payer Name']) + " " + str(column['Plan Name']) +" "+str(column['Insurance Type']))
  ecw_insurance.append(str(column['ECW Insurance']))
  ids.append(column['Ids'])
print(ecw_insurance)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Sentences we want to encode. Example:
ins_mapping_embeddings = []
for field in mapping_fields:
  embeds = model.encode(field)
  ins_mapping_embeddings.append(embeds.tolist())
# Sentences are encoded by calling model.encode()
print(ins_mapping_embeddings)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
entities = [ids,ecw_insurance, ins_mapping_embeddings]
```

```python
insert_result = collection.insert(entities)
# After final entity is inserted, it is best to call flush to have no growing segments left in memory
collection.flush()
```

```python
index_params = {"metric_type":"COSINE", "index_type":"IVF_FLAT", "params":{"nlist":1024}, "index_name":"'eligibility_insurance_attribute_index"}
```

```python
collection.create_index(field_name="eligibility_insurance_attribute_vector", index_params=index_params)
```

```python
collection.create_index(
  field_name="ecw_insurance_name",
  index_name="scalar_index",
  index_params = {"index_type":"marisa-trie"}
)
collection.load()
```

```python
query = 'Blue Cross Community MMAI (Medicare-Medicaid Plan) Blue Cross Community Health Plans Medicaid'
query_embeddings = model.encode(query)
```

```python
embd = query_embeddings.tolist()
```

```python
collection = Collection("ECWEligibilityInsurances")
result = collection.search(
    data=[embd],
    anns_field='eligibility_insurance_attribute_vector',
    param={"metric_type":"L2", "params":{"nlist":1024}},
    limit=5,
    expr=None,
    output_fields=['ecw_insurance_name']


)
```
