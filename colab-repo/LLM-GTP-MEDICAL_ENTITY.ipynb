```python


from openai import OpenAI
import os
from datetime import datetime

os.environ['OPENAI_API_KEY'] = '1234'
client = OpenAI()

t1 = datetime.now()
response = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "user",
      "content": query
    }
  ],
  temperature=1,
  max_tokens=1000,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
t2 = datetime.now()
print("Elapsed time :",  (t2-t1).total_seconds())
```
